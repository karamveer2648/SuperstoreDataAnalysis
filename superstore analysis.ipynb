```python
#Sales Trend Analysis on the basis of region
Salestrendanalysis = df.groupby('Region').sum()['Sales'].sort_values(ascending=False)
Salestrendanalysis.plot(kind='bar',color='red',title='Sales Trend Analysis on the basis of region')
```

```python
#According to above data we can say that West region has highest sales and South region has lowest sales
#According to above data we can say that Central region has second highest sales and East region has second lowest sales
#This helps us to understand the sales trend in different regions
```

```python
#Sales Trend Analysis on the basis of Category
SalesCategory = df.groupby('Category').sum()['Sales'].sort_values(ascending=False)
SalesCategory.plot(kind='bar',color='blue',title='Sales Trend Analysis on the basis of Category')
```

```python
#According to above data we can say that Technology category has highest sales and Furniture category has lowest sales
#According to above data we can say that Office Supplies category has second highest sales
#This helps us to understand the sales trend in different categories
```

```python
#Top 5 states in terms of sales
Top5states = df.groupby('State').sum()['Sales'].sort_values(ascending=False).head(5)
Top5states.plot(kind='bar',color='green',title='Top 5 states in terms of sales')
```

```python
#According to above data we can say that California has highest sales and New York has second highest sales
#According to above data we can say that Texas has third highest sales and Washington has fourth highest sales
#According to above data we can say that Pennsylvania has fifth highest sales
#This helps us to understand the sales trend in different states
```

```python
#Top and Bottom 5 cities in terms of sales
Top5cities = df.groupby('City').sum()['Sales'].sort_values(ascending=False).head(5)
Top5cities.plot(kind='bar',color='green',title='Top 5 cities in terms of sales')
```

```python
#According to above data we can say that New York City has highest sales and Los Angeles has second highest sales
#According to above data we can say that Seattle has third highest sales and San Francisco has fourth highest sales
#According to above data we can say that Philadelphia has fifth highest sales
#This helps us to understand the sales trend in different cities
```

```python
#Bottom 5 states in terms of sales
bottom5states = df.groupby('State').sum()['Sales'].sort_values(ascending=True).head(5)
bottom5states.plot(kind='bar',color='orange',title='Bottom 5 states in terms of sales')
```

```python
#According to above data we can say that North Dakota has lowest sales and West Virginia has second lowest sales
#According to above data we can say that Maine has third lowest sales and South Dakota has fourth lowest sales
#According to above data we can say that Wyoming has fifth lowest sales
#This helps us to understand the sales trend in different states
```

```python
#Bottom 5 cities in terms of sales
bottomcities = df.groupby('City').sum()['Sales'].sort_values(ascending=True).head(5)
bottomcities.plot(kind='bar',color='orange',title='Bottom 5 cities in terms of sales')
```

```python
#According to above data we can say that Abilene has lowest sales and Elyria has second lowest sales
#According to above data we can say that Romeoville has third lowest sales and Littleton has fourth lowest sales
#According to above data we can say that Port Orange has fifth lowest sales
#This helps us to understand the sales trend in different cities
```

```python
#Profit Trend Analysis on the basis of ship mode
Profitshipmode = df.groupby('Ship Mode').sum()['Profit'].sort_values(ascending=False)
Profitshipmode.plot(kind='bar',color='red',title='Profit Trend Analysis on the basis of ship mode')
```

```python
#According to above data we can say that Standard Class ship mode has highest profit and Second Class ship mode has lowest profit
#According to above data we can say that First Class ship mode has second highest profit and Same Day ship mode has second lowest profit
#This helps us to understand the profit trend in different ship modes
```

```python
#analysis of customer segment according to sales
SegmentSales = df.groupby('Segment').sum()['Sales'].sort_values(ascending=False)
SegmentSales.plot(kind="pie",title='Analysis of customer segment according to sales',autopct='%1.1f%%')
```

```python
#According to above data we can say that Consumer segment has highest sales and Home Office segment has lowest sales
#According to above data we can say that Corporate segment has second highest sales
#This helps us to understand the sales trend in different segments
```

```python
#analysis of customer segment according to profit
SegmentProfit = df.groupby('Segment').sum()['Profit'].sort_values(ascending=False)
SegmentProfit.plot(kind='pie',color='blue',title='Analysis of customer segment according to profit',autopct='%1.1f%%')
```

```python
#According to above data we can say that Consumer segment has highest profit and Home Office segment has lowest profit
#According to above data we can say that Corporate segment has second highest profit
#This helps us to understand the profit trend in different segments
```
